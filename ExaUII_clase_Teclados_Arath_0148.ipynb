#### La clase Teclado es un ejemplo de cómo la programación puede representar objetos físicos del mundo real mediante la abstracción.

* Identidad de los Atributos: En este modelo, nos enfocamos en las propiedades esenciales que definen a un periférico: la marca (fabricante), el tipo de switch (mecanismo de presión) y el idioma del layout (configuración de teclas).

* Interactividad Autónoma: Gracias a las funciones leerDatos() y mostrarDatos(), el objeto es capaz de gestionar su propio flujo de información. Esto significa que el código principal se vuelve más limpio, ya que solo le "pide" al objeto que realice sus tareas.

* Modularidad: Al crear dos objetos independientes (teclado1 y teclado2), el programa permite comparar diferentes configuraciones de hardware sin que los datos de un teclado interfieran con los del otro, demostrando la eficiencia y el orden de la estructura por clases.
![texto del vínculo](https://raw.githubusercontent.com/JacoboValenzuela128/ExaUII-Clase-Teclado-POO-Arath-0148/refs/heads/main/infoexaIIclassTeclado.png)

```python
class Teclado:
    def __init__(self, marca="", tipoSwitch="", idiomaLayout=""):
        self.marca = marca
        self.tipoSwitch = tipoSwitch
        self.idiomaLayout = idiomaLayout

    def leerDatos(self):
        print(f"\n⌨️ Registrando nuevo periférico...")
        self.marca = input("Marca del teclado: ")
        self.tipoSwitch = input("Tipo de switch (ej. Blue, Red, Brown, Membrana): ")
        self.idiomaLayout = input("Idioma del layout (ej. Español, Inglés): ")

    def mostrarDatos(self):
        print(f"\n--- Especificaciones del Teclado: {self.marca} ---")
        print(f"🔘 Switches: {self.tipoSwitch}")
        print(f"🌎 Distribución: {self.idiomaLayout}")
        print("-" * 40)

# Creación de los 2 objetos (Instancias)

teclado1 = Teclado()
teclado2 = Teclado()

# Ejecución de las funciones

# Proceso para el primer teclado
teclado1.leerDatos()
teclado1.mostrarDatos()

# Proceso para el segundo teclado
teclado2.leerDatos()
teclado2.mostrarDatos()

print("\n-------------------------------------------------------------------")
print(" Programa interactivo creado por Arath Valenzuela (NL52)")
```
